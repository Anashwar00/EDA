```python
import pandas as pd
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
```

```python
df = pd.DataFrame()
def api_request():
    global df
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
    'start':'1',
    'limit':'15',
    'convert':'USD'
}
    headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': '6b1024a9-bb03-450d-8840-f768fb9a2e8d',
}

    session = Session()
    session.headers.update(headers)

    try:
        response = session.get(url, params=parameters)
        data = json.loads(response.text)
        print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
        print(e)
    df2=pd.json_normalize(data['data'])
    
    df2['Timespamp']=pd.to_datetime('now')
    df = df._append(df2)
    if not os.path.isfile(r"C:\Users\jessy pinkman\Documents\crypto2.csv"):
        df.to_csv(r"C:\Users\jessy pinkman\Documents\crypto2.csv",header='column_names')
    else:
        df.to_csv(r"C:\Users\jessy pinkman\Documents\crypto2.csv",mode='a',header=False)
        
    
```

```python
df1=df.groupby('symbol')[['quote.USD.percent_change_1h','quote.USD.percent_change_24h','quote.USD.percent_change_7d','quote.USD.percent_change_30d','quote.USD.percent_change_60d','quote.USD.percent_change_90d']].mean()
```
